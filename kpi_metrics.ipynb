```python
def ticket_breakdown(df):
    breakdown=df['Status'].value_counts().reset_index()
    return breakdown
    
    
def total_tickets(df):
    return df['Status'].count()
    
def average_response_time(df):
    df_average_response_time = df[df['Created'].notna() & df['Updated'].notna()]
    df_average_response_time['Updated_Timestamp'] = pd.to_datetime(df_average_response_time['Updated'])
    df_average_response_time['Created_Timestamp'] = pd.to_datetime(df_average_response_time['Created'])
    df_average_response_time['Resolution_Time'] = df_average_response_time['Updated_Timestamp'] - df_average_response_time[
        'Created_Timestamp']
    average_response_time = math.ceil(df_average_response_time['Resolution_Time'].mean().total_seconds() / (24 * 3600))
    return average_response_time

def col1_col2_breakdown(df,col1,col2):
    df_request_type = df[~(df[col1].isnull() & df[col2].isnull())]
    df_request_type[col1] = df_request_type[col1].str.strip()
    df_request_type[col2] = df_request_type[col2].str.strip()
    # request_type_count = df_request_type['Request_Type'].value_counts()
    # #region_count.to_dict()
    # request_category_count = df_request_type['Request_Category'].value_counts()
    # # department_count.to_dict()
    col1_type_col2_type = (df_request_type
                         .groupby([col1, col2])
                         .size()
                         .reset_index()
                         )
    return col1_type_col2_type

def assignnee_workload(df):
    workload_df=df[df['Assignee'].notnull()]
    workload=workload_df['Assignee'].value_counts()
    return workload
    
def top_performers(df):
    performer_df=df[df['Assignee'].notnull()]
    performer_df=(performer_df[performer_df['Status'].isin(['Completed','Closed'])]
                  .groupby(['Assignee'])
                  .agg({'Resolution_Time':"mean",'Issue_key': "count"})
                  .reset_index()
                  )
    # Convert timedelta to total days
    performer_df['Avg_Resolution_Days'] = performer_df['Resolution_Time'].dt.total_seconds() / (24 * 3600)
    
    # Compute custom performance score
    performer_df['Performance_Score'] = performer_df['Issue_key'] / performer_df['Avg_Resolution_Days']
    
    # Sort by performance score (higher = better)
    performer_df = performer_df.sort_values(by='Performance_Score', ascending=False)
    return performer_df

# Define the sentiment function
def analyze_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity
def classify_sentiment(p):
    if p > 0.1:
        return 'Positive'
    elif p < -0.1:
        return 'Negative'
    else:
        return 'Neutral'
def clean_comment(text):
    if pd.isna(text) or not isinstance(text, str):
        return ""

    # Normalize base noise
    text = text.replace("nan", " ")

    # Remove email addresses and URLs
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'https?://\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove dates, timestamps, and IDs
    text = re.sub(r'\d{2}/\d{2}/\d{4}', '', text)
    text = re.sub(r'\d{2}/[A-Za-z]{3}/\d{4}', '', text)
    text = re.sub(r'\b\d{1,2}:\d{2}\s*(AM|PM|am|pm)?', '', text)
    text = re.sub(r'\b[a-f0-9]{24}\b', '', text)

    # Remove tracking codes or Jira references
    text = re.sub(r'\[[A-Z]+-\d+\]', '', text)
    text = re.sub(r'\[IT-\d+\|?.*?\]', '', text)

    # Remove salutations and signatures
    text = re.sub(r'^(hi|hello|dear)\b.*?(?=\n)', '', text, flags=re.IGNORECASE | re.MULTILINE)
    text = re.sub(r'thanks.*|regards.*|sincerely.*|best.*|cheers.*', '', text, flags=re.IGNORECASE)

    # Remove system-generated templates
    system_patterns = [
        r'This is an automated reminder.*',
        r'You are receiving this email.*',
        r'Ticket migrated.*',
        r'Powered by Jira.*',
        r'Begin forwarded message.*',
        r'DH changed the status.*',
        r'Ticket From - Partner Support.*?status of \*.*?\*'
    ]
    for pattern in system_patterns:
        text = re.sub(pattern, '', text, flags=re.IGNORECASE)

    # Remove markdown, unicode, and special formatting
    text = re.sub(r'{[^}]+}', '', text)
    text = re.sub(r'\[.*?\|.*?\]', '', text)
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    text = re.sub(r'[*_“”"–—]', '', text)

    # Collapse excess newlines, spaces, dashes
    text = re.sub(r'[-_]{2,}', ' ', text)
    text = re.sub(r'\s+', ' ', text)

    return text.strip()

# def clean_comment(comment):
#     if comment is None or not isinstance(comment, str):
#         return "", []  # Return empty string and empty list if no comment text
#     
#     # 1. Extract Jira references (e.g., [PROJ-1234]) using regex for patterns like PROJECT-1234
#     jira_ids = re.findall(r'\b[A-Z]+-\d+\b', comment)
#     # Deduplicate while preserving order
#     jira_ids = list(dict.fromkeys(jira_ids))
#     
#     # 2. Remove system markers and irrelevant boilerplate phrases
#     comment = re.sub(r'^\s*qm:\s*', '', comment, flags=re.IGNORECASE)                # remove "qm:" prefixes
#     comment = re.sub(r'\(\d+\s*kB\)', '', comment, flags=re.IGNORECASE)              # remove attachment size like "(44 kB)"
#     comment = re.sub(r'h6\. Message originally posted.*?on [AP]M ', '', comment, flags=re.DOTALL)  
#     comment = re.sub(r'h6\. Message originally posted by in .*?on [AP]M ', '', comment, flags=re.DOTALL)
#     comment = re.sub(r'Please check this issue for more details', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'We will reply as soon as possible\.?', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'We will respond to your request in a timely manner during business hours\. '
#                      r'You can expect a response by .*?business day for requests sent outside of business hours\.', 
#                      '', comment, flags=re.IGNORECASE|re.DOTALL)
#     comment = re.sub(r'We will look into this\.?', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Working on it', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'This should get moved to ERP', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'This is an internal note', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'The linked issue\s*-\s*has been resolved', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'We hope this email finds you well\.?', '', comment, flags=re.IGNORECASE)
#     # Remove business hours info blocks
#     comment = re.sub(r'between Memorial Day and Labor Day are Monday through Thursday, 8 a\.m\. to 5 p\.m and Friday 8a?m to 3 p\.m\.', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'are Monday through Friday, 8 a\.m\. to 5 p\.m\.', '', comment, flags=re.IGNORECASE)
#     
#     # 3. Remove greetings and names
#     comment = re.sub(r'\bGood (?:morning|afternoon|evening),\s*', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'\bHi\s+[^,\n]+[,!]', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'\bHello\s+[^,\n]+[,!]', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'\bDear\s+[^,\n]+[,!]', '', comment, flags=re.IGNORECASE)
#     # Remove standalone name if it remains at start (e.g., "John." from "Good morning, John.")
#     comment = re.sub(r'^[A-Z][a-z]+\.\s+', '', comment)
#     
#     # 4. Remove courtesy phrases and personal sign-offs
#     comment = re.sub(r'Thank you for contacting customer (?:support|service)\.?', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Thank you, Customer Relations Capital Area Food Bank', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Customer Relations Capital Area Food Bank', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Thanks, [^-\n]+- Partner Support', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Thank you for your response!?', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Thank you\s*(?:so\s+much|very\s+much)?!?', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Thanks for [^.]+\.', '', comment, flags=re.IGNORECASE)
#     comment = re.sub(r'Be blessed!?', '', comment, flags=re.IGNORECASE)
#     # Remove trailing or standalone "thank you"
#     comment = re.sub(r'thank you!?$', '', comment, flags=re.IGNORECASE)
#     # Remove quoted initials (e.g., "D" from sign-offs)
#     comment = re.sub(r'"\s*[A-Z]\s*"', '', comment)
#     
#     # 5. Clean up extra punctuation and whitespace from removals
#     comment = comment.replace('\\', '')                        # remove any leftover backslashes
#     comment = re.sub(r'\.\s*,', '.', comment)                  # fix period followed by comma
#     comment = re.sub(r'-{2,}', ' ', comment)                   # replace long dashes or "----" with space
#     comment = re.sub(r'^[\s,\.!?]+', '', comment)              # strip leading whitespace or punctuation
#     comment = re.sub(r'[,\.\-!?]+\s*$', '', comment)           # strip trailing punctuation
#     comment = re.sub(r'\s+,', ',', comment)                    # remove space before commas
#     comment = re.sub(r'\s+\.', '.', comment)                   # remove space before periods
#     comment = re.sub(r'\s+!', '!', comment)                    # remove space before exclamation
#     comment = re.sub(r'\s+', ' ', comment)                     # collapse multiple spaces
#     comment = comment.strip()
#     
#     # jira_ids - contains jira tickets per issue
#     return comment



def generate_comment_sentiments(df):
    comment_cols = [col for col in df.columns if col.startswith("Comment")]
    null_comment_df=(df.dropna(subset=comment_cols,how='all'))
    null_comment_df['Consumer_Comments']=null_comment_df[comment_cols].astype(str).agg(' '.join, axis=1).apply(clean_comment)
    return null_comment_df

def generate_polarity_subjectivity_for_comments(df):
    df[['Polarity', 'Subjectivity']] = df['Consumer_Comments'].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)
    df['Sentiment_Label'] = df['Polarity'].apply(classify_sentiment)
    return df

def breakdown_of_sentiments(df):


    ##Sentiment comment breakdown
    
    # Total comments
    # total_comments = null_comment_df.shape[0]
    
    # Sentiment label distribution
    sentiment_counts = df['Sentiment_Label'].value_counts().to_dict()
    
    # Polarity/subjectivity averages
    avg_polarity = df['Polarity'].mean()
    avg_subjectivity = df['Subjectivity'].mean()
    return( sentiment_counts, avg_polarity, avg_subjectivity)

def weekly_sentiments(df):
    df['Created'] = pd.to_datetime(df['Created'], errors='coerce')
    df = df.dropna(subset=['Created'])
    
    # Weekly sentiment trend
    weekly_sentiment = df.set_index('Created').resample('W').agg({
        'Polarity': 'mean',
        'Sentiment_Label': lambda x: x.value_counts().to_dict()
    }).reset_index()
    return weekly_sentiment

def sentiment_by_dept(df):
    ###Sentiment Anlaysis by department

    # Avg polarity per department
    polarity_by_dept = df.groupby('Relevant_Departments')['Polarity'].mean().sort_values().reset_index()
    
    # Count of comments per department
    count_by_dept = df['Relevant_Departments'].value_counts().reset_index()
    
    # # Display
    # print(polarity_by_dept)
    # print(count_by_dept)
    
    return (polarity_by_dept,count_by_dept)

def top_k_negative_comments(df,k=10):
    ##most negative comments 
    negative_comments = df[df['Sentiment_Label'] == 'Negative']
    
    # Sort by polarity (most negative first)
    top_negatives = negative_comments.sort_values(by='Polarity').head(k)
    
    top_negatives[['Created', 'Consumer_Comments', 'Polarity']]
    return top_negatives

def word_cloud_generation(df):
    # Generate from negative comments
    neg_text = ' '.join(df['Consumer_Comments'].dropna().astype(str))
    
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(neg_text)
    
    return wordcloud

```

```python


def forecast_weekly_ticket_volume(adf, weeks_ahead=4):
    ts = adf[['Created_Timestamp']].dropna().copy()
    ts['Created_Timestamp'] = pd.to_datetime(ts['Created_Timestamp'])
    ts.set_index('Created_Timestamp', inplace=True)
    weekly_counts = ts.resample('W').size()

    model = ARIMA(weekly_counts, order=(1,1,1))
    fit = model.fit()
    forecast = fit.forecast(steps=weeks_ahead)

    # Combine actual and forecast into a single DataFrame
    combined = pd.concat([
        weekly_counts.rename('observed'),
        forecast.rename('forecast')
    ])
    return combined.reset_index().rename(columns={'index': 'Week'})


def forecast_weekly_resolution_time(adf, weeks_ahead=4):
    ts = adf[['Created_Timestamp', 'Resolution_Time']].dropna().copy()
    ts['Created_Timestamp'] = pd.to_datetime(ts['Created_Timestamp'])
    ts['Resolution_Days'] = ts['Resolution_Time'].dt.total_seconds() / (24 * 3600)

    weekly_avg = ts.groupby(pd.Grouper(key='Created_Timestamp', freq='W'))['Resolution_Days'].mean()

    model = ARIMA(weekly_avg, order=(1,1,1))
    fit = model.fit()
    forecast = fit.forecast(steps=weeks_ahead)

    combined = pd.concat([
        weekly_avg.rename('observed'),
        forecast.rename('forecast')
    ])
    return combined.reset_index().rename(columns={'Created_Timestamp': 'Week'})

def backlog_trend(adf):
    """
    Returns a DataFrame showing weekly cumulative ticket inflow, resolved count, backlog, and backlog change.
    """

    # Drop nulls for both timestamps
    df = adf[['Created_Timestamp', 'Resolved_Timestamp']].dropna(how='all').copy()

    # Set to weekly buckets
    df['Created_Week'] = pd.to_datetime(df['Created_Timestamp']).dt.to_period('W').dt.start_time
    df['Resolved_Week'] = pd.to_datetime(df['Resolved_Timestamp']).dt.to_period('W').dt.start_time

    # Weekly ticket creation
    created_weekly = df['Created_Week'].value_counts().sort_index().cumsum().rename('Cumulative_Created')

    # Weekly resolution
    resolved_weekly = df['Resolved_Week'].value_counts().sort_index().cumsum().rename('Cumulative_Resolved')

    # Combine both into a timeline
    backlog_df = pd.concat([created_weekly, resolved_weekly], axis=1).fillna(method='ffill').fillna(0)

    # Calculate backlog and weekly change
    backlog_df['Backlog'] = backlog_df['Cumulative_Created'] - backlog_df['Cumulative_Resolved']
    backlog_df['Backlog_Change'] = backlog_df['Backlog'].diff().fillna(0)

    # Reset index for output
    backlog_df = backlog_df.reset_index().rename(columns={'index': 'Week'})

    return backlog_df

def weekly_inflow_outflow(adf):
    """
    Returns weekly ticket inflow (created) vs outflow (resolved) counts.
    """

    df = adf[['Created_Timestamp', 'Resolved_Timestamp']].dropna(how='all').copy()

    df['Created_Week'] = pd.to_datetime(df['Created_Timestamp']).dt.to_period('W').dt.start_time
    df['Resolved_Week'] = pd.to_datetime(df['Resolved_Timestamp']).dt.to_period('W').dt.start_time

    inflow = df['Created_Week'].value_counts().sort_index().rename('Inflow')
    outflow = df['Resolved_Week'].value_counts().sort_index().rename('Outflow')

    combined = pd.concat([inflow, outflow], axis=1).fillna(0).reset_index().rename(columns={'index': 'Week'})

    return combined

def percent_negative_comments_over_time(sentiment_df):
    """
    Returns a weekly breakdown of percentage of negative comments.
    Expects sentiment_df to have columns: Created, Sentiment_Label
    """

    df = sentiment_df[['Created', 'Sentiment_Label']].copy()
    df['Created'] = pd.to_datetime(df['Created'], errors='coerce')
    df = df.dropna(subset=['Created'])

    df['Week'] = df['Created'].dt.to_period('W').dt.start_time

    # Count total and negative comments per week
    weekly_counts = df.groupby('Week').size().rename("Total_Comments")
    weekly_negatives = df[df['Sentiment_Label'] == 'Negative'].groupby('Week').size().rename("Negative_Comments")

    # Merge and compute %
    combined = pd.concat([weekly_counts, weekly_negatives], axis=1).fillna(0)
    combined['% Negative'] = (combined['Negative_Comments'] / combined['Total_Comments']) * 100

    return combined.reset_index()

def negative_ticket_resolution_time_trend(adf, sentiment_df):
    """
    Returns average resolution time (in days) for negatively labeled tickets per week.
    """

    # Step 1: Filter negative sentiment tickets with valid Created and Resolved dates
    df = sentiment_df[sentiment_df['Sentiment_Label'] == 'Negative'].copy()
    df = df[['Issue_key', 'Created']].dropna()

    merged = pd.merge(df, adf[['Issue_key', 'Created_Timestamp', 'Resolution_Time']], on='Issue_key', how='inner')
    merged = merged.dropna(subset=['Created_Timestamp', 'Resolution_Time'])

    # Step 2: Calculate resolution in days
    merged['Resolution_Days'] = merged['Resolution_Time'].dt.total_seconds() / (24 * 3600)
    merged['Week'] = merged['Created_Timestamp'].dt.to_period('W').dt.start_time

    # Step 3: Weekly average resolution time
    result = (
        merged.groupby('Week')['Resolution_Days']
        .mean()
        .reset_index()
        .rename(columns={'Resolution_Days': 'Avg_Resolution_Negative_Tickets'})
    )

    return result

def escalated_tickets_per_agent_trend(adf):
    """
    Returns a weekly count of escalated tickets (e.g., Priority == High) assigned to each agent.
    """
    df = adf[(adf['Priority'].notna()) & (adf['Priority'].str.lower() == 'high') & (adf['Assignee'].notna())].copy()
    
    df['Week'] = df['Created_Timestamp'].dt.to_period('W').dt.start_time

    result = (
        df.groupby(['Week', 'Assignee'])
        .size()
        .reset_index(name='Escalated_Ticket_Count')
    )

    return result

```

```python


def build_ticket_json(row):
    row['Consumer_Comments']=clean_comment(row['Consumer_Comments'])
    extracted = extract_fields(row['Consumer_Comments'])

    return {
        "request_id": row.get('Issue_key', ""),
        "created_at": row.get('Created', ""),
        "customer_message": row.get('Consumer_Comments', ""),
        "intent": extracted.get("intent", ""),
        "target": extracted.get("target", ""),
        "timeframe": extracted.get("timeframe", ""),
        "status": row.get("Status", ""),
        "department": row.get("Relevant_Departments", ""),
        "assignee": row.get("Assignee", "")
    }
from transformers import T5Tokenizer, T5ForConditionalGeneration
import json

# Load FLAN-T5
model_name = "google/flan-t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

def extract_fields(text):
    prompt = f"""
### Task
You are an AI assistant for support teams. Extract structured intent from real customer messages only.

### Instruction
From the message below, extract:
- intent (what the user wants)
- target (what the issue or object is)
- timeframe (any date, urgency or time window)

If the message has no real customer content (like system messages), return empty strings.

### Format
Respond only in this JSON format:
{{"intent": "...", "target": "...", "timeframe": "..."}}

### Examples
Input: I need to reschedule my pantry visit from Saturday to next week.
Output: {{"intent": "reschedule appointment", "target": "pantry visit", "timeframe": "next week"}}

Input: Can someone help me submit my SNAP renewal today?
Output: {{"intent": "submit renewal", "target": "SNAP", "timeframe": "today"}}

Input: This is an automated reminder. Linked issue has been resolved.
Output: {{"intent": "", "target": "", "timeframe": ""}}

### Message
{text.strip()}

### Output
""".strip()

    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    output = model.generate(input_ids, max_length=100, num_beams=4, early_stopping=True)
    raw = tokenizer.decode(output[0], skip_special_tokens=True)

    try:
        fields = json.loads(raw)
    except json.JSONDecodeError:
        fields = {"intent": "", "target": "", "timeframe": ""}

    return fields

def summarize_comment_agentic(comment):
    """
    Breaks the comment into readable bullet points.
    """
    if not isinstance(comment, str) or not comment.strip():
        return []

    from nltk.tokenize import sent_tokenize
    sentences = sent_tokenize(comment)

    # Filter vague/system lines
    filtered = [s for s in sentences if len(s.split()) >= 4 and "posted in" not in s.lower() and "ticket migrated" not in s.lower()]

    # Return cleaned bullets
    return [f"- {s.strip()}" for s in filtered[:6]]  # Limit to 6 bullet points

def estimate_urgency(comment):
    """
    Checks for urgency cues in the text.
    """
    if not isinstance(comment, str): return 0

    comment = comment.lower()
    urgent_terms = ['urgent', 'asap', 'immediately', 'today', 'tomorrow', 'by end of day', 'this week']
    urgency_score = sum(term in comment for term in urgent_terms)

    return min(urgency_score, 3)  # Cap at 3

def estimate_seriousness(comment):
    from textblob import TextBlob

    if not isinstance(comment, str): return 0

    blob = TextBlob(comment)
    polarity = blob.sentiment.polarity

    severity_keywords = ['missing', 'broken', 'issue', 'error', 'not working', 'problem']
    score = sum(k in comment.lower() for k in severity_keywords)

    seriousness = score + (1 if polarity < -0.3 else 0)

    return min(seriousness, 3)  # Cap at 3

def compute_priority_score(urgency, seriousness):
    """
    Assigns a final priority level based on urgency + seriousness
    """
    score = urgency + seriousness
    if score >= 5:
        return "High"
    elif score >= 3:
        return "Medium"
    else:
        return "Low"

def agentic_summary_pipeline(row):
    comment = row.get("Consumer_Comments", "")

    summary = summarize_comment_agentic(comment)
    urgency = estimate_urgency(comment)
    seriousness = estimate_seriousness(comment)
    priority = compute_priority_score(urgency, seriousness)

    return pd.Series({
        "bullet_points": summary,
        "urgency": urgency,
        "seriousness": seriousness,
        "priority": priority
    })


```
